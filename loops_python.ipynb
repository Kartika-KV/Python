# **Decision making**
Decision making allows us to run a particular block of code after validating a particular conditions.


Condition checking is the backbone of decision making.

# **1. The if statement**
The if statement is used to test a particular condition and if the condition is true, a block of code (if-block) will be executed.

The condition of if statement can be any valid logical expression which can be either evaluated to true or false.

# **if Statement Syntax**

if test expression:

      statement(s)  # if-block or body of if-block

# **Note:**
In Python, the body of the if statement is indicated by the indentation. The body starts with an indentation and the first unindented line marks the end.

# **Example:**

```python
#print positive number
num = -1
if num > 0:
    print(num, "is a positive number.")
print("This is always printed.")
```

# **2. Python if...else Statement**
The if-else statement provides an else block combined with the if statement which is executed in the false case of the condition.

If the condition is true, then the if-block is executed. Otherwise, the else-block is executed.


# **Syntax of if...else**
if test expression:

     Body of if
else:

      Body of else

# **Python if..else Flowchart**

image.png

# **Example of if...else**

```python
# Program checks if the number is positive or negative
# And displays an appropriate message

num = 3

# Try these two variations as well. 
num = -5
num = 0

if num > 0:
    print("Positive or Zero")
else:
    print("Negative number")
```

# **3. if...elif...else Statement**

The elif statement enables us to check multiple conditions and execute the specific block of statements depending upon the true condition among them.

 We can have any number of elif statements in our program depending upon our need.

 

# **Syntax of if...elif...else**
if test expression:

      Body of if
elif test expression:

      Body of elif
else: 

     Body of else

Only one block among the several if...elif...else blocks is executed according to the condition.

The if block can have only one else block. But it can have multiple elif blocks.

# **Flowchart of if...elif...else**
image.png

# **Example of if...elif...else**

```python
'''In this program, 
we check if the number is positive or
negative or zero and 
display an appropriate message'''

num = 3

# Try these two variations as well:
num = 0
num = -4

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

# **Nested if statements**

We can have a if...elif...else statement inside another if...elif...else statement. This is called nesting in computer programming.

Any number of these statements can be nested inside one another. Indentation is the only way to figure out the level of nesting. They can get confusing, so these must be avoided unless necessary.

# **Python Nested if Example**



```python
'''In this program, we input a number
check if the number is positive or
negative or zero and display
an appropriate message
This time we use nested if statement'''

num = float(input("Enter a number: "))
if num >= 0:
    if num == 0:
        print("Zero")
    else:
        print("Positive number")
else:
    print("Negative number")
```

# **Why we use loops in python?**
The looping simplifies the complex problems into the easy ones. It enables us to alter the flow of the program so that instead of writing the same code again and again, we can repeat the same code for a finite number of times. 

1. It provides code re-usability.
2. Using loops, we do not need to write the same code again and again.
3. Using loops, we can traverse over the elements of data structures. 

For example, if we need to print the first 10 natural numbers then, instead of using the print statement 10 times, we can print inside a loop which runs up to 10 iterations.

# **1. for loop**

The for loop in Python is used to iterate the statements or a part of the program several times. 

It is frequently used to traverse the data structures like list, tuple, or dictionary. 

# **Syntax of for Loop**
for val in sequence:

	  Body of for

Here, val is the variable that takes the value of the item inside the sequence on each iteration.

Loop continues until we reach the last item in the sequence. 

The body of for loop is separated from the rest of the code using indentation.

# **Example: Python for Loop**


```python
# example
# Program to find the sum of all numbers stored in a list

# List of numbers
numbers = [6, 5, 3, 8, 4, 2, 5, 4, 11]
# sum = numbers[0]+mubers[1].........
# variable to store the sum
sum = 0

# iterate over the list
for val in numbers:
  print(val)
  # val = 6
  sum = sum+val
  #print("The sum1 is", sum)
print(numbers)
print("The sum is", sum)
```

# **The range() function**
We can generate a sequence of numbers using range() function. 

range(start,stop,step size) #default start is 0, step size is 1.

```python
print(range(10)) #The range object is "lazy" in a sense because it doesn't generate every number that it "contains" when we create it. 

print(list(range(10)))

print(list(range(2, 8)))

print(list(range(2, 20, 3)))
```

We can use the range() function in for loops to iterate through a sequence of numbers.

 It can be combined with the len() function to iterate through a sequence using indexing.

```python
# Program to iterate through a list 

fruits = ['apple', 'orange', 'banana']

# iterate over the list using index
for i in fruits: #i = element
	print("I like", i)
```

```python
# Program to iterate through a list using indexing

fruits = ['apple', 'orange', 'banana']

# iterate over the list using index
for i in range(len(fruits)): #i = index
	print("I like", fruits[i])
```

# **Nested for loop**
Python allows us to nest any number of for loops inside a for loop. 

The inner loop is executed n number of times for every iteration of the outer loop.

# **Syntax for nested for loops**



```python
for iterating_var1 in sequence:  #outer loop  
    for iterating_var2 in sequence:  #inner loop  
        #block of statements     
#Other statements  
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

# **2. While loop**

The while loop in Python is used to iterate over a block of code as long as the test expression (condition) is true. 

We generally use this loop when we don't know the number of times to iterate beforehand.

# **Syntax of while Loop**

while test_expression:

      Body of while

# **Flowchart of while Loop**
image.png

# **Example: Python while Loop**

```python
# Another Example:Program to add natural numbers up to n
# sum = 1+2+3+...+n

# To take input from the user,
# n = int(input("Enter n: "))

n = 10

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1    # update counter

# print the sum
print("The sum is", sum)
```

# Loop Control Statements
We can change the normal sequence of while/for loop's execution using the loop control statement.
1. 
break Statement - When the break statement is encountered, it brings control out of the loop.

2. continue Statement - When the continue statement is encountered, the control transfer to the beginning of the loop.

 The continue statement skips the remaining lines of code inside the loop and start with the next iteration.
