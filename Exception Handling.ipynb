# **EXCEPTION HANDLING**

An exception is a runtime error which can be handled by the programmer.

All exception are represented as classes in python.

TYPE OF EXCEPTION:-

**Build-in Exception:-** Exception which are already availble in python language. The base class for all build-in exception is Base Exception Class.

**User Defined Exception:-** A programmer can create his own exception is called user defined exception.

Why we need exception handling?


TWO TYPE OF STATEMENT:-
1. Normal Statement :- which will not give any error
2. Critical Statement:- where you have doubt that statement can give you error.

**Try:-** The try block contains code whic may causes exceptions.
Syntax:

    try:
      statement


**Except:-** The except block is used to catch an exception that is raised in the try block. There can be multiple except block for try block .

Syntax:

    except:
       statement

**Else:-** This block will get executed when no exception is raised. Else block is executed after try block.

Syntax:

    else:
       statement

**Finally:-** This block will get executed irresepctive of whether there is exception or not.

Syntax:

    finally:
      statement

# **User Defined Exception:-**

A programmer can create his own exceptions, called user-defined exception or Custom Exception.

Three step follow to make this exception:-

1. Creating Exception Class using Exception Class as Base Class.

2. Raising Exception

3. Handling Exception 

**Creating Exception:-**We can create our own exception by creating a sub class to build-in exception class.

    class MyException(Exception):
       pass



    class MyException(Exception):
       def __init__(self.arg):
          self.msg= arg


** Raising Exception:-** raise statement is used to raise to user defined exception.
    
    raise MyException("Message")

**Handling Exception:-** using try and except block program can handle exception 

    try:
      statement
    except MyException as mye:
      statement

```python
class BalanceException(Exception):
  pass
def checkBalance():
  money = 10000
  withdraw = int(input("Enter the value how much money you want to withdraw"))

  try:
    balance = money - withdraw
    if(balance<=2000):
      raise BalanceException('Insufficient Balance')
    print(balance)
  except BalanceException as be:
    print(be)

checkBalance()
```
